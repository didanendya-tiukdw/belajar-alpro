```python
def fac_int(n):
    fak = 1
    for i in range (1,n+1):
        fak = fak * i
    return fak

print(fac_int(5))
```

```python
#iteratif fibo 

def fibo(n):
    f1,f2, fibon = 1,1,0
    if n==1 or n==2:
        fibon = 1
    else:
        for i in range (1,1+n):
            fibon = f1 + f2
            f1,f2 = f2,fibon
    return fibon

print(fibo(50))
```

```python
#fibo_rekursif
def fibo_r(n):
    if(n==1):
        return 1
    elif (n==2):
        return 1
    else:
        return fibo_r(n-1) + fibo_r(n-2)

print(fibo_r(50))    
```
