```python
filename = input("nama file: ")
handle = open(filename)
c = 0
for line in handle:
    if line.find("ac.uk") != -1:
        c += 1  
    print("Web dengan domain 'ac.uk' ditemukan di \"" + line.strip()+ "\"")

print("ditemukan sebanyak : ",c) 
```

```python
filename = input("nama file: ")
handle = open(filename)
c = 0
for line in handle:
    if line.startswith("Subject"):
        c += 1
        line = line.strip().title()
        print(line)

print("ditemukan sebanyak : ",c) 
```

```python
filename = input("nama file: ")
try:
    handle = open(filename)
    c = 0
    for line in handle:
        c += len(line)
        kb = c / 1000

    print("Ukuran: " + str(kb) + " KB")

except:
    print("File tidak ditemukan")
```
